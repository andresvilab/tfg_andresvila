```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Dataset characteristics
data = {
    'Dataset': ['DDSM', 'MIAS', 'BCDR', 'INbreast', 'CBIS-DDSM'],
    'Number of Images': [11560, 322, 3703, 410, 1644],
    'Digital/Film': ['Film', 'Film', 'Film', 'Digital', 'Digital'],
    'Annotations': ['Basic', 'Basic', 'Detailed', 'Detailed', 'Detailed'],
    'Resolution': ['Variable', '1024x1024', 'Variable', 'High', 'Variable'],
    'FAIR Compliance': ['Low', 'Low', 'High', 'High', 'High']
}

df = pd.DataFrame(data)

print(df)

```

## Number of images

```python
# Bar chart for number of images
plt.figure(figsize=(10, 6))
plt.bar(df['Dataset'], df['Number of Images'], color='skyblue')
plt.xlabel('Dataset')
plt.ylabel('Number of Images')
plt.title('Number of Images in Each Dataset')
plt.show()
```

## FAIR Compliance

```python
# Assign numerical values for FAIR compliance
df['FAIR Score'] = df['FAIR Compliance'].map({'Low': 1, 'Medium': 2, 'High': 3})

# Bar chart for FAIR compliance
plt.figure(figsize=(10, 6))
plt.bar(df['Dataset'], df['FAIR Score'], color='lightgreen')
plt.xlabel('Dataset')
plt.ylabel('FAIR Compliance Score')
plt.title('FAIR Compliance of Each Dataset')
plt.show()
```

### Pie Charts for Digital vs. Film

```python
# Pie chart for digital vs. film
digital_film_counts = df['Digital/Film'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(digital_film_counts, labels=digital_film_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])
plt.title('Digital vs. Film Datasets')
plt.show()
```

## Detailed Analysis and Visualization

```python
# Plotting resolution quality
resolution_quality = {
    'Dataset': ['DDSM', 'MIAS', 'BCDR', 'INbreast', 'CBIS-DDSM'],
    'Resolution Quality': [2, 1, 2, 3, 2]  # Assuming higher number is better quality
}
resolution_df = pd.DataFrame(resolution_quality)

plt.figure(figsize=(10, 6))
sns.barplot(x='Dataset', y='Resolution Quality', data=resolution_df, palette='viridis')
plt.xlabel('Dataset')
plt.ylabel('Resolution Quality Score')
plt.title('Resolution Quality of Each Dataset')
plt.show()
```
